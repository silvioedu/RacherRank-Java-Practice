## Sierpinki Triangle



1st Transformation:
```
x n+1 = 0.5xn
y n+1 = 0.5yn
```

2nd Transformation:
```
x n+1 = 0.5xn + 0.5
y n+1 = 0.5yn + 0.5
```

3rd Transformation:
```
x n+1 = 0.5xn + 1
y n+1 = 0.5yn
```


````
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline

print('Setup complete')
````

````
def transformation_1(p):
  x = p[0]
  y = p[1]
  x1 = 0.5 * x
  y1 = 0.5 * y
  return x1, y1

def transformation_2(p):
  x = p[0]
  y = p[1]
  x1 = 0.5 * x + 0.5
  y1 = 0.5 * y + 0.5
  return x1, y1

def transformation_3(p):
  x = p[0]
  y = p[1]
  x1 = 0.5 * x + 1
  y1 = 0.5 * y
  return x1, y1

transformations = [transformation_1, transformation_2, transformation_3]
````

````
# Initializations
a1 = [0]
b1 = [0]
a, b = 0, 0
````

````
for i in range(1000000):
  transf = choice(transformations)
  a, b = transf((a, b))
  a1.append(a)
  b1.append(b)
````

````
plt.rc('figure', figsize=(16,16))
plt.plot(a1, b1, 'o')
````
